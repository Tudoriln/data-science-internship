```python
data['player_tags'] = data['player_tags'].fillna('').apply(
    lambda x: x.split(', ') if isinstance(x, str) and x.strip() else []
)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
tags_encoded = pd.DataFrame(mlb.fit_transform(data['player_tags']), columns = mlb.classes_, index = data.index)
```

```python
data = data.drop(columns = 'player_tags', errors = 'ignore')
data = pd.concat([data, tags_encoded], axis=1)
data.head()
```

```python
data['player_traits'] = data['player_traits'].fillna('')
data['num_traits'] = data['player_traits'].apply(
    lambda x: len([trait.strip() for trait in x.split(',') if trait.strip()]) if x else 0
)
```

```python
data['body_type'] = data['body_type'].apply(lambda x: 0 if x == 'Lean' else 1 if x == 'Normal' else 2)
```

```python
pd.set_option('display.max_rows', 200)  # > 162
print(data['nationality'].value_counts())
```

```python
threshold_percent = 0.5
total_players = data.shape[0]
threshold_value = total_players * (threshold_percent / 100)

#count nationalities
nationality_counts = data['nationality'].value_counts()

#identify common and rare countries
common_nationalities = nationality_counts[nationality_counts >= threshold_value].index
rare_nationalities = nationality_counts[nationality_counts < threshold_value].index

#build the encoding map for common nationalities
encoding_map = {name: idx for idx, name in enumerate(common_nationalities)}

#Assign rare countries a common value
rare_code = 999
data['nationality'] = data['nationality'].apply(lambda x: encoding_map.get(x, rare_code))
#the last line of code works like an if inside a loop (for): if checks if the country it finds is rare or not and either add the rare_code or the corect value from the map
data.head()
```

```python
data[["Atk_effort", "Def_effort"]] = data['work_rate'].str.split(pat = '/', n=1, expand = True)
data = data.drop('work_rate', axis=1)
```

```python
data['Atk_effort'].value_counts()
data['Atk_effort'] = data['Atk_effort'].apply(lambda x: 0 if x == 'Low' else 1 if x == 'Medium' else 2)
#data.head()
```

```python
data['Def_effort'].value_counts()
data['Def_effort'] = data['Def_effort'].apply(lambda x: 0 if x == 'Low' else 1 if x == 'Medium' else 2)
```

```python
data['preferred_foot'].value_counts()
data['preferred_foot'] = data['preferred_foot'].apply(lambda x: 1 if x == 'Right' else 0)
```

```python
pd.set_option('display.max_rows', 200)  # > 162
print(data['team_position'].value_counts())
```

```python
def map_position(pos):
    if pos == 'SUB':
        return 'SUB_position'
    elif pos == 'RES':
        return 'RES_position'
    elif pos == 'GK':
        return 'GK_position'
    elif pos in ['ST', 'CF', 'LF', 'RF', 'RS', 'LS', 'RW', 'LW']:
        return 'Att_position'
    elif pos in ['CM', 'RCM', 'LCM', 'CAM', 'CDM', 'RDM', 'LDM', 'LAM', 'RAM', 'RM', 'LM']:
        return 'Mid_position'
    elif pos in ['CB', 'RCB', 'LCB', 'RB', 'LB', 'RWB', 'LWB']:
        return 'Def_position'
```

```python
data['position_group'] = data['team_position'].apply(map_position)

#One-hot encode
encoded_positions = pd.get_dummies(data['position_group'])

data = pd.concat([data, encoded_positions], axis=1)
data.drop(columns=['position_group'], inplace=True)
```

```python
col1 = data['release_clause_eur']
col2 = data['value_eur']

#Calculate scaling factor
k = np.sum(col1 * col2) / np.sum(col1 * col1)
data['release_clause_eur'] = data['value_eur'].fillna(data['release_clause_eur'] / k)
```

```python
def sum_parts(val):
    parts = val.split('+')
    parts = [int(p) for p in parts if p.strip().isdigit()]
    return sum(parts)

data['ls'] = data['ls'].apply(sum_parts)
data['rs'] = data['rs'].apply(sum_parts)
data['lw'] = data['lw'].apply(sum_parts)
data['lf'] = data['lf'].apply(sum_parts)
data['cf'] = data['cf'].apply(sum_parts)
data['rf'] = data['rf'].apply(sum_parts)
data['rw'] = data['rw'].apply(sum_parts)
data['lam'] = data['lam'].apply(sum_parts)
data['cam'] = data['cam'].apply(sum_parts)
data['ram'] = data['ram'].apply(sum_parts)
data['lm'] = data['lm'].apply(sum_parts)
data['lcm'] = data['lcm'].apply(sum_parts)
data['cm'] = data['cm'].apply(sum_parts)
data['rcm'] = data['rcm'].apply(sum_parts)
data['rm'] = data['rm'].apply(sum_parts)
data['lwb'] = data['lwb'].apply(sum_parts)
data['ldm'] = data['ldm'].apply(sum_parts)
data['cdm'] = data['cdm'].apply(sum_parts)
data['rdm'] = data['rdm'].apply(sum_parts)
data['rwb'] = data['rwb'].apply(sum_parts)
data['lb'] = data['lb'].apply(sum_parts)
data['lcb'] = data['lcb'].apply(sum_parts)
data['cb'] = data['cb'].apply(sum_parts)
data['rcb'] = data['rcb'].apply(sum_parts)
data['st'] = data['st'].apply(sum_parts)
data['rb'] = data['rb'].apply(sum_parts)

```
