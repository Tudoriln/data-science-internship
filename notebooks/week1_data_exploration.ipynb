```python
data.drop_duplicates(inplace = True)
#nu exista duplicate in datele raw
```

```python
counts = data["body_type"].value_counts()
data = data[data["body_type"].isin(counts[counts > 1].index)]
```

```python
data.head()
```

```python
data = data.reset_index(drop=True)
```

```python
data['player_tags'] = data['player_tags'].fillna('').apply(
    lambda x: x.split(', ') if isinstance(x, str) and x.strip() else []
)
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
tags_encoded = pd.DataFrame(mlb.fit_transform(data['player_tags']), columns = mlb.classes_, index = data.index)
```

```python
data = data.drop(columns = 'player_tags', errors = 'ignore')
data = pd.concat([data, tags_encoded], axis=1)
data.head()
```

```python
import numpy as np
```

```python
gk_cols = ['gk_diving', 'gk_handling', 'gk_kicking', 'gk_reflexes', 'gk_speed', 'gk_positioning']
outfield_cols = ['pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic']
data['player_rating'] = np.where(
    data['team_position'] == 'GK', data[gk_cols].mean(axis=1),
        data[outfield_cols].mean(axis=1)
)
```

```python
data['player_rating'] = data['player_rating'].round().astype('Int64')
```

```python
data['player_traits'] = data['player_traits'].fillna('')
data['num_traits'] = data['player_traits'].apply(
    lambda x: len([trait.strip() for trait in x.split(',') if trait.strip()]) if x else 0
)
```
